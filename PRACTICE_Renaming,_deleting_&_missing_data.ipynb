```python
# importing library
import pandas as pd
```

```python
# loading data sheet
filename = '/content/drive/MyDrive/data/RailsToTrails_National CountData_week.xlsx'
bike_df = pd.read_excel(filename)
bike_df
```

```python
#delting 7,8,9
bike_df = bike_df.drop(columns=['Unnamed: 7',	'Unnamed: 8',	'Unnamed: 9'])
```

```python
bike_df.head()
```

```python
# renaming the Unnamed: 6 column to percent_change
bike_df = bike_df.rename(columns={'Unnamed: 6': 'percent_change'})
bike_df.head()
```

```python
# finding missing valyes and columns
bike_df.loc[0:10,:].dropna(how = 'all')
```
