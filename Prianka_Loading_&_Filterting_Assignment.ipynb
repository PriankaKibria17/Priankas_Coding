```python
# selecting data after July 31, 2020
weekof_filter = bike_df['Week of'] > '2020-07-31'
```

```python
# selecting data where the change 2019-2020 is greater than 100% (greater than 1)
change_filter = bike_df['Change 2019-2020'] > 1
```

```python
#Putting together/combining
#  Answer to question 3: No. There is nothing that is both after 7/31 and has a change >1
bike_df = bike_df.loc[weekof_filter & change_filter, :]
bike_df   
```
