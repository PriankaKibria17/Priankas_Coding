```python
# letting python know this data does not have a header
df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data', header = None)
```

```python
#using option 1 to rename columns
## in order to save the columns you have to give it a name or use inpace = True
df = df.rename(columns = {0: 'Flow_ID',
                     1: 'Date',
                     2: 'Time',
                     3: 'Count'})
```

```python
#renaming columns option 2
df.columns = ['Flow_ID',
              'Date',
              'Time',
              'Count']
df.head()
```

```python
# data for July 24 2005 having flow id=7
# get the date 7/24/05 and the flow id
# when you use the & inbetween you have to use parenthesis
data = (df['Date'] == '07/24/05') & (df['Flow_ID'] == 7)
```

```python
df[data].head()
```

```python
# to get an idea of the amount of data there is 
len(df)
```

```python
# to get the amount of data you got
len(df[data])
```

```python
# sorting results on count
#output the top ten rows
#assign the output ot a variable named top10
```

```python
#looking at my data 
df.loc[data, :].head()
```

```python
# you can use the same data above to sort it now in descending order & output the top ten rows
# sorting my valyes by using sort_values and mentioning by = ['Count'] 
# since i want it in descending order I will use ascending = False
# noting I want the number of data I want is ten .head(10)
# saving it as a top ten variable
top_ten = df.loc[data, :].sort_values(by = ['Count'], ascending = False).head(10)
top_ten
```

```python
# using panda apply function to subtract lowest value of the 10 from all of them
# then output the average value of the resulting counts
# lowest value is 1, will subtract 1 from tge rest in column count

#accessing the lowest value
lowestVal = top_ten['Count'].min()
lowestVal
```

```python
# looking at the count row series
top_ten['Count']
```

```python
# using apply function
# allows me to use a function on all the data in this specific column
# lambda function is an unamed function (argument inside of a function)
# subtracting 1 - lowest val
top_ten['Count'].apply(lambda x: x - lowestVal)
```

```python
# getting the average using .mean
top_ten['Count'].apply(lambda x: x - lowestVal).mean()
```

```python
# grouping
# selecting data in the month of Aug 2005 having flow id=7

# adjusting my prev data 
#using starts with method and ends with

data = (df['Flow_ID'] == 7) & (df['Date'].str.startswith('08')) & (df['Date'].str.endswith('05'))
```

```python
# grouping data based on date and getting the max count per date

#only locating the rows where data is true

df.loc[data, :].groupby('Date').max()
```
